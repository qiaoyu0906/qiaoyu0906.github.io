```python
from PIL import Image

# 打开原始图像
original_image = Image.open("stage_exm.PNG")

# 步骤1: 降低分辨率 (将图像尺寸缩小)
downsampled_image = original_image.resize((original_image.width // 4, original_image.height // 4), Image.NEAREST)

# 步骤2: 降低颜色数 (颜色量化)
# 将图像转换为调色板模式，并强制使用较少的颜色（如16种颜色）
downsampled_image = downsampled_image.convert("P", palette=Image.ADAPTIVE, colors=8)

# 为了更接近你展示的效果，可以进一步修改调色板来去除特定的颜色
palette = downsampled_image.getpalette()

# 将调色板的蓝色（背景色）修改为黑色
for i in range(0, len(palette), 3):
    r, g, b = palette[i:i+3]
    if r > 200 and g > 200 and b > 200:  # 检测白色或接近白色
        palette[i:i+3] = [0, 0, 0]  # 设置为黑色

downsampled_image.putpalette(palette)

# 如果你想显示图像，先转换回RGB模式
downsampled_image_rgb = downsampled_image.convert("RGB")

# 保存降采样后的图像
downsampled_image_rgb.save("downsampled_image_custom.png")

# 展示降采样后的图像
downsampled_image_rgb.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from mpl_toolkits.mplot3d import Axes3D

def plot_reward_surface(filename,plotname):
    # 读取 CSV 文件
    df = pd.read_csv(filename)

    # 从 CSV 文件中提取数据
    theta = df['theta'].values
    d = df['d'].values
    Z = df['Z'].values

    # 创建规则网格以进行插值
    theta_grid, d_grid = np.meshgrid(np.linspace(0, 1, 100), np.linspace(-0.1, 0.1, 100))

    # 对数据进行插值以生成表面
    Z_grid = griddata((theta, d), Z, (theta_grid, d_grid), method='cubic')

    # 创建 3D 曲面图
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(111, projection='3d')

    # 绘制插值后的表面图
    surf = ax.plot_surface(theta_grid, d_grid, Z_grid, cmap='coolwarm', edgecolor='none')

    ax.set_xlabel('Parameter')
    ax.set_ylabel('Perturbation')
    ax.set_zlabel('Reward')

    plt.title(f'{plotname}')
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()

# 使用函数绘制三张图
filenames = [
    '存在局部高峰theta_d_z_data.csv',
    '存在较多局部最优theta_d_z_data.csv',
    '整体平缓theta_d_z_data.csv'
]
plotnames=[
  'Reward Surface of A function in 3.4.2.1',
    'Reward Surface of A function in 3.4.2.2',
    'Reward Surface of B function in 3.4.2.2'
]
for filename,plotname in zip(filenames,plotnames):
    plot_reward_surface(filename,plotname)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_heatmap(image_path, heatmap_data=None, block_size=10):
    # 读取图片
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 如果没有提供热力图数据，生成一个随机的热力图
    if heatmap_data is None:
        # 生成更大区域的随机块
        heatmap_data = np.random.rand(img.shape[0] // block_size, img.shape[1] // block_size)
        heatmap_data = cv2.resize(heatmap_data, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)

    # 将热力图数据标准化到0-255
    heatmap_data = cv2.normalize(heatmap_data, None, 0, 255, cv2.NORM_MINMAX)
    heatmap_data = np.uint8(heatmap_data)

    # 将热力图数据转换为彩色图像
    heatmap = cv2.applyColorMap(heatmap_data, cv2.COLORMAP_JET)

    # 叠加热力图到原始图像上
    superimposed_img = cv2.addWeighted(img, 0.7, heatmap, 0.3, 0)

    # 显示结果
    plt.figure(figsize=(10, 8))
    plt.imshow(superimposed_img)
    plt.axis('off')
    plt.show()

# 使用该函数，设定较大的块大小
apply_heatmap('4.png', block_size=20)
```

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt
from torchvision.models import resnet50
from torchcam.methods import SmoothGradCAMpp
from PIL import Image

# 定义变换，适合预训练模型的输入
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# 读取图像并应用变换
image_path = '4.png'  # 修改为你的图像路径
img = Image.open(image_path).convert('RGB')
input_tensor = transform(img).unsqueeze(0)

# 使用预训练的ResNet50模型
model = resnet50(pretrained=True)
model.eval()

# 使用 torch-cam 的 SmoothGradCAM++ 生成 CAM
cam_extractor = SmoothGradCAMpp(model)
# 选择要可视化的目标类别（可以选择模型输出最高的类别）
outputs = model(input_tensor)
target_class = outputs.argmax().item()

# 生成热力图
activation_maps = cam_extractor(target_class, outputs)

# 处理第一个激活图
activation_map = activation_maps[0]  # 提取列表中的第一个元素
activation_map = activation_map.squeeze().cpu().numpy()

# 将热力图调整为原始图像的大小
heatmap = cv2.resize(activation_map, (img.width, img.height))

# 将热力图标准化为0到255之间
heatmap = cv2.normalize(heatmap, None, 0, 255, cv2.NORM_MINMAX)
heatmap = np.uint8(heatmap)

# 将热力图转换为彩色
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

# 叠加热力图到原始图像上
superimposed_img = cv2.addWeighted(np.array(img), 0.6, heatmap, 0.4, 0)

# 显示结果
plt.figure(figsize=(10, 8))
plt.imshow(superimposed_img)
plt.axis('off')
plt.show()
```
